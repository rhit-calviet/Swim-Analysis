```python
import os
import pandas as pd
from pandas.errors import EmptyDataError

# Define the base path where the directories are located
base_path = 'C:\\Users\\ranadeva\\Downloads'

# Initialize an empty list to store dataframes
dataframes = []

# Iterate through each directory
for dir_name in os.listdir(base_path):
    dir_path = os.path.join(base_path, dir_name)

    # Check if it is a directory
    if os.path.isdir(dir_path):
        # Determine gender from the directory name
        if dir_name.startswith('Women'):
            gender = 'F'
            division_index = 6  # For 'WomenDX_YY_ZZ', 'D' is at index 5, 'X' is at index 6
        else:
            gender = 'M'
            division_index = 4  # For 'MenDX_YY_ZZ', 'D' is at index 3, 'X' is at index 4

        # Determine division from the directory name
        division_mapping = {'1': 'I', '2': 'II', '3': 'III'}
        division = division_mapping[dir_name[division_index]]

        # Determine season from the directory name
        season = dir_name.split('_')[1] + '_' + dir_name.split('_')[2]

        # Iterate through each CSV file in the directory
        for file_name in os.listdir(dir_path):
            if file_name.endswith('.csv'):
                file_path = os.path.join(dir_path, file_name)

                try:
                    # Read the CSV file into a DataFrame
                    df = pd.read_csv(file_path)

                    # Add the new columns
                    df['Gender'] = gender
                    df['Division'] = division
                    df['Season'] = season

                    # Append the dataframe to the list
                    dataframes.append(df)

                except EmptyDataError:
                    print(f"Skipped empty file: {file_path}")

# Concatenate all dataframes if any valid dataframes are found
if dataframes:
    final_df = pd.concat(dataframes, axis=0)

    # Save the concatenated dataframe to a new CSV file
    final_df.to_csv('combined_data.csv', index=False)

    print("All CSV files have been processed and combined.")
else:
    print("No valid data found in any CSV files.")
```
