```python
import pandas as pd

# Assuming 'hs' is the original DataFrame
# Reset the index so 'ID' becomes a column
hs_reset = hs.reset_index()

# Melt the DataFrame to create a long format
melted = hs_reset.melt(
    id_vars=['ID', 'Name', 'School', 'Gender'], 
    value_vars=[col for col in hs.columns if 'Time' in col],
    var_name='Event', 
    value_name='Time'
)

# Filter out rows where 'Time' is <NA>
melted = melted.dropna(subset=['Time'])

# Extract the corresponding points columns for each event
def get_points_column(event):
    return event.replace('Time', 'Points')

# Add a 'Points' column by mapping the event to its points
melted['Points'] = melted.apply(lambda row: hs.at[row['ID'], get_points_column(row['Event'])], axis=1)

# Create the new DataFrame with the required columns
new_df = melted[['ID', 'Name', 'School', 'Event', 'Time', 'Gender', 'Points']]

# Display the new DataFrame
new_df.head()
```

```python
def parse_time_to_timedelta(time_str):
    if '.' in time_str:
        # Case 1: S.CC, SS.CC, M:SS.CC, MM:SS.CC formats
        if ':' in time_str:
            minutes, seconds = time_str.split(':')
            seconds, centiseconds = seconds.split('.')
            return pd.Timedelta(minutes=int(minutes), seconds=int(seconds), milliseconds=int(centiseconds)*10)
        else:
            # Cases like S.CC or SS.CC
            seconds, centiseconds = time_str.split('.')
            return pd.Timedelta(seconds=int(seconds), milliseconds=int(centiseconds)*10)
    return pd.Timedelta(seconds=0)

# Apply the function to the Time column to create the Time_Adj column
new_df['Time_Adj'] = new_df['Time'].apply(parse_time_to_timedelta)
```

```python
# Define a function to modify the 'Event' column
def format_event(row):
    # Map for strokes and courses
    stroke_map = {
        'Free': 'Freestyle',
        'Fly': 'Butterfly',
        'Breast': 'Breaststroke',
        'Back': 'Backstroke',
        'IM': 'Individual Medley'
    }
    
    course_map = {
        'Y': 'SCY',  # Short Course Yard
        'L': 'LCM',  # Long Course Meter
        'S': 'SCM'   # Short Course Meter
    }
    
    # Replace stroke and course abbreviations with full names
    stroke = row['Event'].split()[2]  # Get the stroke from the event string
    course = row['Event'].split()[1]  # Get the course from the event string
    distance = row['Event'].split()[0]  # Get the distance
    gender = 'Male' if row['Gender'] == 'M' else 'Female'  # Set gender
    
    # Apply the mappings
    stroke_longform = stroke_map.get(stroke, stroke)  # Default to original if not found
    course_longform = course_map.get(course, course)  # Default to original if not found
    
    # Construct the new event string
    new_event = f"{distance} {stroke_longform} {course_longform} {gender}"
    
    return new_event

# Apply the function to the 'Event' column
new_df['Event'] = new_df.apply(format_event, axis=1)

# Display the updated DataFrame
new_df.head()
```
