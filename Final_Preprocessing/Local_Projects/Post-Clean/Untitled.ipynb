```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from IPython.display import clear_output

def get_teams_for_swimmer(swimmer_id):
    url = f"https://www.swimcloud.com/swimmer/{swimmer_id}/"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36',
        'Referer': 'https://google.com/'
    }

    while True:
        response = requests.get(url, headers=headers)
        
        if response.status_code != 200:
            print(f"Failed to retrieve data for swimmer_id {swimmer_id}")
            time.sleep(60)
            return []
        
        soup = BeautifulSoup(response.text, 'html.parser')
        teams_section = soup.find_all('div', class_='c-card c-card--large c-card--divided')[1]
        if not teams_section:
            print("Teams section not found, skipping swimmer...")
            return []
        
        teams_list = teams_section.find_all('h3', class_='c-list__title')
        break

    teams = [team.get_text(strip=True) for team in teams_list]
    return teams

def match_teams_to_schools(df):
    i = 0
    for index, row in df.iterrows():
        swimmer_id = row['ID']
        school = row['School']
        teams = get_teams_for_swimmer(swimmer_id)
        
        matched_team = None
        for team in teams:
            if school.lower() in team.lower() or team.lower() in school.lower():
                matched_team = team
                break
        
        if matched_team:
            print(f"Match found for {row['Name']}: {matched_team}")
            df.drop(index, inplace=True)  # Drop the row from DataFrame
        else:
            print(f"No match found for {row['Name']}: {school}")

        i += 1
        if (i % 10 == 0):
            print(i)
        if (i % 100 == 0):
            clear_output()

        # time.sleep(20)

# Example usage
# dups = pd.read_csv('your_csv_file.csv')  # Load your DataFrame
match_teams_to_schools(dups)
```
