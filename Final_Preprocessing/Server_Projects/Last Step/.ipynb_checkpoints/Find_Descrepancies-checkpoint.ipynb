```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from IPython.display import clear_output

def get_teams_for_swimmer(swimmer_id):
    url = f"https://www.swimcloud.com/swimmer/{swimmer_id}/"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36',
        'Referer': 'https://google.com/'
    }

    while True:
        response = requests.get(url, headers=headers)
        
        if response.status_code != 200:
            print(f"Failed to retrieve data for swimmer_id {swimmer_id}")
            time.sleep(60)
            return []
        
        soup = BeautifulSoup(response.text, 'html.parser')
        teams_section = soup.find_all('div', class_='c-card c-card--large c-card--divided')[1]
        if not teams_section:
            print("Teams section not found, skipping swimmer...")
            return []
        
        teams_list = teams_section.find_all('h3', class_='c-list__title')
        break

    teams = [team.get_text(strip=True) for team in teams_list]
    return teams

def match_teams_to_schools(df1, df):
    i = 0
    count = 0
    for index, row in df1.iterrows():
        swimmer_id = row['ID']
        school = row['School']
        teams = get_teams_for_swimmer(swimmer_id)
        
        matched_team = None
        for team in teams:
            if school.lower() in team.lower() or team.lower() in school.lower():
                matched_team = team
                break
        
        if matched_team:
            print(f"Match found for {row['Name']}: {matched_team}")
            # Update 'School' in df directly for matching ID and Name
            df.loc[(df['ID'] == row['ID']) & (df['Name'] == row['Name']), 'School'] = matched_team
            
            # Drop the row from df1 as it's processed
            df1.drop(index, inplace=True)
            count = count + 1
        else:
            print(f"No match found for {row['Name']}: {school}")

        i += 1
        if i % 10 == 0:
            print(i)
        if i % 100 == 0:
            clear_output()

    # Drop rows where all elements are NaN
    df.dropna(how='all', inplace=True)
    print(count)

# Example usage
# dups has ID, Name, School
# dups = tr.drop('Power',axis=1)[['ID','Name','School']]
dups = pd.read_csv('1_21_25_correcteddups.csv')
df.School = df.School.astype('string')
match_teams_to_schools(dups, df)
```
