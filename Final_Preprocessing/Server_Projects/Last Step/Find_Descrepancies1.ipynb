```python
import pandas as pd
```

```python
df = pd.read_parquet('CLEANED_SWIM_DATA.parquet')
df.head(2)
```

```python
school_counts = df.groupby('Name')['School'].nunique()

# Filter for names with more than one school
names_with_multiple_schools = school_counts[school_counts > 1]

# Count how many names have more than one school
len(names_with_multiple_schools)
```

```python
tr = df[df['Name'].isin(names_with_multiple_schools.index)]
tr.shape
```

```python
tr.loc[:,'isHS'] = tr.loc[:,'Division'] == 'HS'
tr.head()
```

```python
g = tr.groupby(['ID','isHS'])
g.apply(lambda x:x.head(),include_groups=False)
```

```python
pivot_table = tr.pivot_table(values='Points', index='ID', columns='isHS', aggfunc='mean')

# Add a new column for the difference between False and True averages
pivot_table['Difference'] = abs(pivot_table[False] - pivot_table[True])

# Display the pivot table with the new 'Difference' column
pivot_table
```

```python
pivot_table.sort_values(by='Difference')
```

```python
pivot_table[pivot_table.Difference < 200].sort_values(by='Difference')
```

```python
bad_ids = pivot_table[pivot_table.Difference > 200].sort_values(by='Difference').index
bad_ids
```

```python
df.shape
```

```python
df = df[~df['ID'].isin(bad_ids)]
```

```python
df.shape
```

```python
df.to_parquet('CLEANED_SWIM_DATA.parquet')
```
