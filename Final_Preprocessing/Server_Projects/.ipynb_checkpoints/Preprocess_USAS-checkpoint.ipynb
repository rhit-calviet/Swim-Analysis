```python
def parse_time_to_timedelta(time_str):
    if '.' in time_str:
        # Case 1: S.CC, SS.CC, M:SS.CC, MM:SS.CC formats
        if ':' in time_str:
            minutes, seconds = time_str.split(':')
            seconds, centiseconds = seconds.split('.')
            return pd.Timedelta(minutes=int(minutes), seconds=int(seconds), milliseconds=int(centiseconds)*10)
        else:
            # Cases like S.CC or SS.CC
            seconds, centiseconds = time_str.split('.')
            return pd.Timedelta(seconds=int(seconds), milliseconds=int(centiseconds)*10)
    return pd.Timedelta(seconds=0)

# Apply the function to the Time column to create the Time_Adj column
df['Time_Adj'] = df['Time'].apply(parse_time_to_timedelta)
df.head(2)
```

```python
df['Time_Sec'] = df['Time_Adj'].apply(lambda x: x.total_seconds())
df.head(2)
```

```python
records = pd.read_csv('combined_events.csv')
```

```python
merged_df = pd.merge(df, records[['Event', 'Time']], on='Event', how='left', suffixes=('_df', '_record'))
merged_df.head(2)
```

```python
# Compute the points based on the provided formula
merged_df['points'] = 1000 * (merged_df['Time_record'] / merged_df['Time_Sec'])**3

# If you want to keep the result in 'df', you can assign the 'points' column
df['Points'] = merged_df['points']
df.head(2)
```

```python
# For Div_Rank: Rank within Event, Gender, SeasonYear, and Division
df['Div_Rank'] = df.groupby(['SeasonYear', 'Event', 'Division'], observed=False)['Time_Sec'].rank(method='min', ascending=True)

# For NCAA_Rank: Rank within Event, Gender, and SeasonYear (across all divisions)
df['NCAA_Rank'] = df.groupby(['SeasonYear', 'Event'], observed=False)['Time_Sec'].rank(method='min', ascending=True)
df.head(2)
```
