```python
import pandas as pd
```

```python
df = pd.read_csv('HS_times.csv')
df.head(2)
```

```python
df.info()
```

```python
df.index = df.index.astype('category')
```

```python
for col in df.columns[1:45]:
    df[col] = df[col].astype('string')
```

```python
df.info()
```

```python
df.to_parquet('HS_times_cleaned.parquet',index=True)
```
